### Q5. Functions

```python
def is_prime(n):
    # Prime numbers are greater than 1
    if n <= 1:
        return False
    # Check for factors from 2 to sqrt(n)
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Test the function
print(f"Is 7 prime? {is_prime(7)}")
print(f"Is 15 prime? {is_prime(15)}")
print(f"Is 23 prime? {is_prime(23)}")
print(f"Is 2 prime? {is_prime(2)}")
print(f"Is 1 prime? {is_prime(1)}")
```

### Q6. Lists

```python
numbers = [5, 2, 9, 1, 7, 3]

# 1. Sort the list without using sort()
# Using bubble sort for demonstration
n = len(numbers)
sorted_list = list(numbers) # Create a copy to not modify original
for i in range(n):
    for j in range(0, n-i-1):
        if sorted_list[j] > sorted_list[j+1]:
            sorted_list[j], sorted_list[j+1] = sorted_list[j+1], sorted_list[j]
print(f"Original list: {numbers}")
print(f"Sorted list (without sort()): {sorted_list}")

# 2. Find the sum of all elements
list_sum = 0
for num in numbers:
    list_sum += num
print(f"Sum of all elements: {list_sum}")

# 3. Remove all odd numbers and print the remaining list
even_numbers = []
for num in numbers:
    if num % 2 == 0:
        even_numbers.append(num)
print(f"List after removing odd numbers: {even_numbers}")
```
